```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# read map 
# map_file = "/media/rp/Elements/abhay_ws/real_contact_data/cross_peg_contact_mapping_real_45.csv"
# map_file = "/media/rp/Elements/abhay_ws/mujoco_contact_graph_generation/results/cross_data/perturb/processed_data/cross_peg_contact_map_sim.csv"
# map_file = "/media/rp/Elements/abhay_ws/mujoco_contact_graph_generation/results/extrusion_data/perturb_v2/processed_data/extrusion_peg_contact_map_sim.csv"
map_file = "/home/rp/dhanush_ws/cicp/data/extrusion_real_map.csv"
# map = pd.read_pickle(map_file)
map = pd.read_csv(map_file) 
```

```python
# histogram of map data 
dims = ['X','Y','Z','A','B','C']
# dims = ['x','y','z','a','b','c']
coords = [(0,0), (0,1), (0,2), (1,0), (1,1), (1,2)]

fig, axs = plt.subplots(2, 3, figsize=(15, 10)) 
for dim in dims:
    ax = axs[coords[dims.index(dim)][0], coords[dims.index(dim)][1]] 
    ax.hist(map[dim], bins='auto', label=dim) 
    ax.axvline(np.mean(map[dim]), color='r', linestyle='--', label='Mean') 
    ax.legend() 
    ax.set_title(dim)
plt.show()
```
